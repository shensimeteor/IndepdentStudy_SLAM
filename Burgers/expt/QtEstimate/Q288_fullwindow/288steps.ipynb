```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file="concat_ens_step288.bin"
data=np.fromfile(file)
nx=1000
n_ens=int(data.shape[0]/nx)
data=data.reshape((n_ens, nx))
data.shape
```

```python
cov=np.cov(data.transpose())
print(cov.shape)
cor=np.corrcoef(data.transpose())
print(cor.shape)
```

```python
#print(np.sqrt(np.diag(cov)))
fig=plt.figure(figsize=(12,12))
plt.imshow(cor[:100,:100])
cbar=plt.colorbar(shrink=0.8, orientation='horizontal', pad=0.05)
cbar.ax.tick_params(labelsize=14)
plt.title("Corr. Coeff. Matrix of P(0->288)", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)

#plt.savefig("p288_corrcoef_matrix.png")
```

```python
fig=plt.figure(figsize=(12,12))
plt.plot(cor[400,:])
plt.title("Corr. Coeff. of P(0->288) (w.r.t. x=400 point)", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.savefig("p288_corrcoeff_x400.png")

fig=plt.figure(figsize=(12,12))
plt.plot(np.arange(350,450), cor[400,350:450])
plt.title("Corr. Coeff. of P(0->288) (w.r.t. x=400 point)", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.savefig("p288_corrcoeff_x400_zoomin.png")

fig=plt.figure(figsize=(12,12))
plt.plot(cor[700,:])
plt.title("Corr. Coeff. of P(0->288) (w.r.t. x=700 point)", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.savefig("p288_corrcoeff_x700.png")

fig=plt.figure(figsize=(12,12))
plt.plot(np.arange(650,750), cor[700,650:750])
plt.title("Corr. Coeff. of P(0->288) (w.r.t. x=700 point)", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.savefig("p288_corrcoeff_x700_zoomin.png")
```

```python
fig=plt.figure(figsize=(12,12))
plt.plot(np.sqrt(np.diag(cov)))
plt.title("std. dev. of P (sqrt of P(0->288)' diagonal)", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.savefig("p288_stddev.png")
```

```python
steps=[36,72,108,144,180,216,252,288]
for step in steps:
    file="concat_ens_step%0.3d.bin" %step
    data=np.fromfile(file)
    nx=1000
    n_ens=int(data.shape[0]/nx)
    data=data.reshape((n_ens, nx))
    cor=np.corrcoef(data.transpose())
    fig=plt.figure(figsize=(12,12))
    plt.plot(cor[700,650:750])
    plt.title("Corr. Coeff. of P(0->%0.3d) (w.r.t. x=700 point)" %step, fontsize=24)
    plt.yticks(fontsize=18)
    plt.xticks(fontsize=18)
    plt.savefig("p%0.3d_corrcoeff_x700_zoomin.png"%step)
```
