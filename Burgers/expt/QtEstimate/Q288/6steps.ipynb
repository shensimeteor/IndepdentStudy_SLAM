```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file="concat_ens_step282.bin"
data=np.fromfile(file)
nx=1000
n_ens=int(data.shape[0]/nx)
data=data.reshape((n_ens, nx))
data.shape
```

```python
cov=np.cov(data.transpose())
print(cov.shape)
cor=np.corrcoef(data.transpose())
print(cor.shape)
```

```python
#print(np.sqrt(np.diag(cov)))
fig=plt.figure(figsize=(12,12))
plt.imshow(cor)
cbar=plt.colorbar(shrink=0.8, orientation='horizontal', pad=0.05)
cbar.ax.tick_params(labelsize=14)
plt.title("Corr. Coeff. Matrix of Q6", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)

plt.savefig("q6_corrcoef_matrix.png")
```

```python
fig=plt.figure(figsize=(12,12))
plt.plot(cor[400,:])
plt.title("Corr. Coeff. of Q6 (w.r.t. x=400 point)", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.savefig("q6_corrcoeff_x400.png")
```

```python
fig=plt.figure(figsize=(12,12))
plt.plot(np.sqrt(np.diag(cov)))
plt.title("std. dev. of Q6 (sqrt of Q6' diagonal)", fontsize=24)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.savefig("q6_stddev.png")
```
