```python
#!/usr/bin/env python
import numpy as np
import matplotlib.pyplot as plt
file="output.bin"
nx=1000
N=1441
Xs=np.fromfile(file, dtype="double", count=-1).reshape(N,nx)

dt=0.001 # timestep is 0.001s
xcoord=np.linspace(-np.pi, np.pi, 1000)


```

```python
#t=0.01
i=int(0.01/dt)
plt.figure()
plt.plot(xcoord, Xs[i][:], '-o')
plt.show()
```

```python
#t=1
i=int(1/dt)
plt.figure()
plt.plot(xcoord, Xs[i][:], '-o')
plt.show()
```
