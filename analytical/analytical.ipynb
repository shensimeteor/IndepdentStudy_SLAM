```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y0 = - 0.5*np.pi #u0 is cos(x) for x in [-pi, pi], so y0 = -0.5pi
```

```python
def solution(t, y0, x, miu):
    f=1/t * (x-y0 - np.pi -np.pi*np.tanh((np.pi * (x-y0 - np.pi))/(2*miu*t)))
    return f
    
```

```python
x=np.linspace(-np.pi, np.pi, 1000)
print(x.shape)
```

```python
miu = 1e-4
t=1

f=solution(t, y0, x, miu)
#print(f)
plt.plot(x,f)
plt.title("t=1")
#break
```

```python
t=0.1
f=solution(t, y0, x, miu)
#print(f)
plt.plot(x,f)
plt.title("t=0.1")
```
